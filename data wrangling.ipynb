```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("Dataset.csv")
data.head(5)
```

```python
data.drop('Date/Time',axis=1,inplace=True) #will not affect target variable
```

```python
data.isnull().any()
```

```python
#No missing values in the dataset
import seaborn as sns
sns.heatmap(data.corr(),annot=True)
```

theoretical_power_curve and wind direction corelation value fall in range of -0.099 which is approx eqal to 0,ie no linear relationship b/w target variable and wind direction

```python
data.drop('Wind Direction (°)',axis=1,inplace=True)
```
